````python
#Soal nomor 2, fungsi untuk memeriksa sebuah angka apakah prima atau tidak
def prime_checker(number):
    if number == 1 or number == 2:
        if number == 2:
            print(f"{number} is prime number")
        else:
            print(f"{number} is not prime number")
    elif number > 2:
        x=0
        for num in range(2,number):
            if number%num==0:
                x = 1
        if x == 1:
            print(f"{number} is not prime Number")
        else:
            print(f"{number} is prime number")
    else:
        raise ValueError("Please input positive integer and more than 1 !")

n = int(input("Check this number: "))
prime_checker(n)
````
