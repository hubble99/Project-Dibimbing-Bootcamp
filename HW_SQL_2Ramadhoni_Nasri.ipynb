```python
%%sql
select
date_part('month', date(order_date)) as month,
avg(omset) as total_omset
from fact_sales
group by month
order by 1 desc
```

```python
%%sql
select dc.gender,
sum(fc.omset) as total_omset
from dim_customer dc
join fact_sales fc on fc.customer_id = dc.customer_id
join dim_product dp on dp.product_id =fc.product_id
where dc.age>30 and dp.category = 'Bikes'
group by dc.gender
```
